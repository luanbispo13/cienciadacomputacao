```python
def quicksort_quiz(lista, nivel=0):
    """
    QuickSort interativo em formato de quiz.
    Mostra pivô e pergunta para o usuário como dividir os elementos.
    """
    if len(lista) <= 1:
        return lista

    indent = "  " * nivel
    pivot = lista[0]
    print(f"\n{indent}Lista atual: {lista}")
    print(f"{indent}Pivô escolhido: {pivot}")

    menores = []
    maiores = []

    for x in lista[1:]:
        resposta = input(f"{indent}O número {x} vai para 'menores' (m) ou 'maiores' (M)? ").strip().lower()
        if resposta == 'm':
            menores.append(x)
        else:
            maiores.append(x)

    print(f"{indent}Menores: {menores}")
    print(f"{indent}Maiores: {maiores}")

    # Recursão para cada sublista
    return quicksort_quiz(menores, nivel + 1) + [pivot] + quicksort_quiz(maiores, nivel + 1)


# --- Código principal ---
print("Quick Sort Interativo")
entrada = input("Digite números separados por espaço: ")
lista = [int(x) for x in entrada.split()]

print("\nVamos ordenar a lista!\n")
lista_ordenada = quicksort_quiz(lista)

print("\nLista ordenada:", lista_ordenada)
```
