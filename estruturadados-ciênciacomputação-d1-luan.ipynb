Manipulação de Listas com Python - Listas - 29.02 ⏰


```python
#criar uma lista com 5 elementos
#em python uma lista é criada com uma variável (nome) seguida do parâmetro []
#bolos é o nome da lista seguida do sinal de "=" e []

bolos=["chocolate", "morango", "baunilha", "banana", "cenoura"]
print(bolos)
```

```python
#método append()
#list.append(x) adiciona um item ao fim da lista

bolos.append("chocolate")
bolos.append("cenoura")
print(bolos)
```

```python
#método remove()
#list.remove(x) remove o primeiro item encontrado na lista cujo valor é igual a x.

bolos.pop(5) #remove o elemento no índice 5 da lista
print(bolos)
```
