##### Questão 1. Apresente em tela (output) toda a base de dados.

```python
import pandas as pd

# Lista de dicionários com os estados e capitais do Brasil
estados_capitais = [
    {"Estado": "Acre", "Capital": "Rio Branco"},
    {"Estado": "Alagoas", "Capital": "Maceió"},
    {"Estado": "Amapá", "Capital": "Macapá"},
    {"Estado": "Amazonas", "Capital": "Manaus"},
    {"Estado": "Bahia", "Capital": "Salvador"},
    {"Estado": "Ceará", "Capital": "Fortaleza"},
    {"Estado": "Espírito Santo", "Capital": "Vitória"},
    {"Estado": "Goiás", "Capital": "Goiânia"},
    {"Estado": "Maranhão", "Capital": "São Luís"},
    {"Estado": "Mato Grosso", "Capital": "Cuiabá"},
    {"Estado": "Mato Grosso do Sul", "Capital": "Campo Grande"},
    {"Estado": "Minas Gerais", "Capital": "Belo Horizonte"},
    {"Estado": "Pará", "Capital": "Belém"},
    {"Estado": "Paraíba", "Capital": "João Pessoa"},
    {"Estado": "Paraná", "Capital": "Curitiba"},
    {"Estado": "Pernambuco", "Capital": "Recife"},
    {"Estado": "Piauí", "Capital": "Teresina"},
    {"Estado": "Rio de Janeiro", "Capital": "Rio de Janeiro"},
    {"Estado": "Rio Grande do Norte", "Capital": "Natal"},
    {"Estado": "Rio Grande do Sul", "Capital": "Porto Alegre"},
    {"Estado": "Rondônia", "Capital": "Porto Velho"},
    {"Estado": "Roraima", "Capital": "Boa Vista"},
    {"Estado": "Santa Catarina", "Capital": "Florianópolis"},
    {"Estado": "São Paulo", "Capital": "São Paulo"},
    {"Estado": "Sergipe", "Capital": "Aracaju"},
    {"Estado": "Tocantins", "Capital": "Palmas"},
    {"Estado": "Distrito Federal", "Capital": "Brasília"}
]

# Convertendo a lista de dicionários em um DataFrame do pandas.
estados_capitais_df = pd.DataFrame(estados_capitais)

# 1. Exibição de toda a base de dados
print("1. Toda a base de dados:")
print(estados_capitais_df)
print()
```

##### Questão 2. Apresente o tamanho do seu dataframe (quantas colunas x linhas)

```python
# O método `shape` retorna uma tupla com o número de linhas e colunas do DataFrame
print("2. Tamanho do dataframe:")
print(estados_capitais_df.shape)
print()
```

##### Questão 3. Acesse a linha (x) e apresente em tela todas as características do item.

```python
# Utiliza `iloc` para acessar a linha no índice 5 (sexta linha já que a contagem começa do zero)
linha = 5
print(f"3. Linha {linha}:")
print(estados_capitais_df.iloc[linha])
print()
```

##### Questão 4. Verifique se o dataframe está vazio.

```python
# O atributo `empty` verifica se o DataFrame está vazio e retorna um valor booleano
print("4. DataFrame está vazio?")
print(estados_capitais_df.empty)
print()
```

##### Questão 5. Apresente em tela os 5 primeiros registros da base de dados.

```python
# Utiliza o método `head` para mostrar as primeiras 5 linhas do DataFrame
print("5. Os 5 primeiros registros:")
print(estados_capitais_df.head())
print()
```

##### Questão 6. Exclua um item (linha) de sua base de dados.

```python
# Utiliza `drop` para remover a linha no índice 10 (11ª linha, já que a contagem começa do zero)
# `reset_index` é usado para resetar os índices do DataFrame após a remoção
estados_capitais_df = estados_capitais_df.drop(10).reset_index(drop=True)
print("6. Base de dados após excluir a linha 10:")
print(estados_capitais_df)
print()
```

##### Questão 7. Adicione um item (linha) na sua base de dados.

```python
# Cria um novo DataFrame com uma única linha contendo um novo estado e capital
# Utiliza `pd.concat` para concatenar o novo DataFrame com o existente
nova_linha = pd.DataFrame([{"Estado": "Novo Estado", "Capital": "Nova Capital"}])
estados_capitais_df = pd.concat([estados_capitais_df, nova_linha], ignore_index=True)
print("7. Base de dados após adicionar nova linha:")
print(estados_capitais_df)
print()
```

##### Questão 8. Transponha a coluna para a linha em sua base de dados.

```python
# Utiliza o método `T` para transpor o DataFrame, trocando linhas por colunas
print("8. Base de dados transposta:")
print(estados_capitais_df.T)
print()
```

##### Questão 9. Apresente em tela somente a 1º e 2º coluna (rótulo) da base de dados.

```python
# Utiliza a indexação de colunas do DataFrame para selecionar e exibir apenas as colunas especificadas
print("9. Apenas as colunas 1 e 2:")
print(estados_capitais_df[['Estado', 'Capital']])
print()
```
