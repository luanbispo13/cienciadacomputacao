```python
import matplotlib.pyplot as plt

# Dados de exemplo
idade = [30, 21, 35, 42, 37, 20, 8, 17, 35, 25]
rendamensal = [4300, 3350, 5200, 4900, 4700, 2100, 1950, 2700, 4000, 4800]

# Crie o gráfico de dispersão com pontos azuis
plt.scatter(idade, rendamensal)

# Adicione rótulos e um título
plt.xlabel('Idade')
plt.ylabel('Renda Mensal')
plt.title('Gráfico de Dispersão')

# Adicione a grade ao gráfico
plt.grid(True)

# Exiba o gráfico
plt.show()
```
