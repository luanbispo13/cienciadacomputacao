```python
# Regra 3

# Complexo ou complicado?

# Faça um programa que informe o número de loops para um contador que imprima todos os números inteiros de 1 até 10 inclusive.

# Usando a função for
for contador in range (1,12): # o for faz o loop dentro de um dataset
  print(contador)
```

```python
# Usando a função while
contador = 0
while contador <= 10:
  print(contador)
  contador = contador + 1 # A cada loop será atríbuido 1 para o contador
```

```python
# Inicializa o contador com o valor 1.
contador = 1

# Inicia um loop while que executa enquanto o contador é maior que 0 e menor que 11.
# Isso garante que o loop execute enquanto o contador estiver dentro do intervalo de 1 a 10.
while contador > 0 and contador < 11:
    # Imprime o valor atual do contador.
    print(contador)

    # Incrementa o contador em 1 para passar para o próximo número.
    contador += 1
```
