```python
import pandas as pd
nomes = ['Ana', 'Bruno','Carla']
idades = [21, 20, 22]
dados = list(zip(nomes, idades))
print(dados)
df = pd.DataFrame(data = dados)
print(df)
```

```python
import pandas as pd
identidade = {'Nome': ['Ana', 'Bruno', 'Carla'],
         'Idade': [21, 20, 22]}
df = pd.DataFrame(data = identidade)
print(df)
```

```python
import pandas as pd
dados = [('Ana', 21), ('Bruno', 20), ('Carla', 22)]
colunas = ['Nome', 'ID']
linhas = ['A', 'B', 'C']
df = pd.DataFrame(data = dados, columns = colunas, index = linhas)
print(df)
print(list(df.index))
print(list(df.columns))
print(df.ndim)
print(df.shape)
print(df.size)
print(df.empty)
```
