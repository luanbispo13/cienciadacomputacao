```python
# Proposta: Dados de Músicas de uma Plataforma de Streaming (Spotify)

# a) Mais Ouvidas
# b) Artista com mais músicas
# c) Gênero mais popular
# d) Música mais recente
# e) Adicionar uma nova coluna com a duração em segundos.

import pandas as pd

# Dados das Músicas
nome_musicas = ['Se Envolve', 'Confissões', 'Uma Wave', 'Romance em Cena', 'Ambiente Errado']
artistas = ['Veigh', 'KayBlack', 'Cabelinho', 'Luisa Sonza', 'Luan Santana']
generos = ['Trap', 'Trap', 'Trap', 'Pop', 'Sertanejo']
ano_lancamento = [2024, 2025, 2026, 2023, 2022]
duracao_minutos = [2.9, 3.3, 3.1, 3.8, 2.2]

# Crie um dicionário contendo os dados das músicas.
dados_musicas = {
    'Música': nome_musicas,
    'Artista': artistas,
    'Gênero': generos,
    'Ano de Lançamento': ano_lancamento,
    'Duração (minutos)': duracao_minutos
}

# Crie um dataframe a partir do dicionário.
musicas_df = pd.DataFrame(dados_musicas)

# a) Mais Ouvidas
print("a) Mais Ouvidas:")
mais_ouvida = musicas_df.loc[musicas_df['Ano de Lançamento'].idxmax()]['Música']
print(mais_ouvida)
print()

# b) Artista com mais músicas
print("b) Artista com mais músicas:")
artista_mais_musicas = musicas_df['Artista'].mode()[0]
print(artista_mais_musicas)
print()

# c) Gênero mais popular
print("c) Gênero mais popular:")
genero_mais_popular = musicas_df['Gênero'].mode()[0]
print(genero_mais_popular)
print()

# d) Música mais recente
print("d) Música mais recente:")
musica_mais_recente = musicas_df.loc[musicas_df['Ano de Lançamento'].idxmax()]['Música']
print(musica_mais_recente)
print()

# e) Adicionar uma nova coluna com a duração em segundos
musicas_df['Duração (segundos)'] = musicas_df['Duração (minutos)'] * 60
print("e) DataFrame de Músicas com Duração em Segundos:")
print(musicas_df)
print()

display(musicas_df)
```
