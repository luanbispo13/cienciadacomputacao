```python
def contar_nomes(lista):
    contagem = {}
    for nome in lista:
        if nome in contagem:
            contagem[nome] += 1
        else:
            contagem[nome] = 1
    return contagem

# Inicializa a lista de nomes vazia
nomes = []

# Loop para adicionar nomes até que o usuário insira uma string vazia
while True:
    nome = input("Digite um nome (ou deixe em branco para encerrar): ")
    if nome == "":
        break
    nomes.append(nome)

# Chamada da função para contar os nomes
resultado = contar_nomes(nomes)

# Exibir o resultado
print("Contagem de nomes:")
for nome, vezes in resultado.items():
    print(f"{nome}: {vezes} vezes")
```

```python
def ocultar_nomes_proprios(mensagem):
    palavras = mensagem.split()
    nova_mensagem = []
    for palavra in palavras:
        if palavra[0].isupper() and palavra.isalpha():
            nova_mensagem.append('*' * len(palavra))
        else:
            nova_mensagem.append(palavra)
    return ' '.join(nova_mensagem)

# Solicita ao usuário que insira a mensagem
mensagem = input("Digite a mensagem: ")

# Chamada da função para ocultar os nomes próprios
nova_mensagem = ocultar_nomes_proprios(mensagem)

# Exibe a nova mensagem
print("Nova mensagem:", nova_mensagem)
```

```python
def produto_escalar(vetor1, vetor2):
    resultado = 0
    for i in range(3):
        resultado += vetor1[i] * vetor2[i]
    return resultado

def produto_vetorial(vetor1, vetor2):
    resultado = [0, 0, 0]
    resultado[0] = vetor1[1] * vetor2[2] - vetor1[2] * vetor2[1]
    resultado[1] = vetor1[2] * vetor2[0] - vetor1[0] * vetor2[2]
    resultado[2] = vetor1[0] * vetor2[1] - vetor1[1] * vetor2[0]
    return resultado

# Solicita ao usuário que insira os valores dos vetores
vetor1 = []
vetor2 = []

print("Insira os valores do primeiro vetor:")
for i in range(3):
    valor = float(input(f"Digite o {i+1}º valor: "))
    vetor1.append(valor)

print("\nInsira os valores do segundo vetor:")
for i in range(3):
    valor = float(input(f"Digite o {i+1}º valor: "))
    vetor2.append(valor)

# Calcula e exibe o produto escalar
escalar = produto_escalar(vetor1, vetor2)
print("\nProduto escalar:", escalar)

# Calcula e exibe o produto vetorial
vetorial = produto_vetorial(vetor1, vetor2)
print("Produto vetorial:", vetorial)
```

```python
# Solicita ao usuário que insira o número N
N = int(input("Digite o número de elementos que deseja inserir na lista: "))

# Inicializa uma lista vazia para armazenar os números
numeros = []

# Loop para inserir N números na lista
for i in range(N):
    numero = float(input(f"Digite o {i+1}º número: "))
    numeros.append(numero)

# Imprime a lista de números
print("Lista de números inseridos:", numeros)

# Calcula e imprime a soma dos itens
soma = sum(numeros)
print("Soma dos itens:", soma)

# Calcula e imprime a média dos valores
media = soma / N
print("Média dos valores:", media)

# Calcula e imprime o maior valor
maior_valor = max(numeros)
print("Maior valor:", maior_valor)

# Calcula e imprime o menor valor
menor_valor = min(numeros)
print("Menor valor:", menor_valor)
```
